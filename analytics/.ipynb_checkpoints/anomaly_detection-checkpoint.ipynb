# ‚ö†Ô∏è Anomaly Detection on Implement Usage
This notebook identifies anomalies in implement usage durations using Z-score and Isolation Forest.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Load usage summary
df = pd.read_csv('../data_pipeline/implement_usage_summary.csv')
df['Duration_min'] = df['Duration_sec'] / 60
df['Start'] = pd.to_datetime(df['Start'])
df['Stop'] = pd.to_datetime(df['Stop'])
```

## üìä Visualize Duration Distribution

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Duration_min'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Implement Usage Durations (min)")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

## üßÆ Anomaly Detection with Z-Score and Isolation Forest

```python
df['z_score'] = (df['Duration_min'] - df['Duration_min'].mean()) / df['Duration_min'].std()
print(df['z_score'].describe())
df['z_anomaly'] = df['z_score'].abs() > 1.5
df[df['z_anomaly']]
```

## üå≤ Anomaly Detection with Isolation Forest

```python
# Drop rows with NaN in duration
df = df.dropna(subset=['Duration_min'])

model = IsolationForest(contamination=0.05, random_state=42)
df['iforest_anomaly'] = model.fit_predict(df[['Duration_min']])
df['iforest_anomaly'] = df['iforest_anomaly'] == -1
df[df['iforest_anomaly']]
```

## üîÅ Compare Anomaly Detection Methods

```python
overlap = df[df['z_anomaly'] & df['iforest_anomaly']]
print(f"Overlap in anomalies: {len(overlap)} out of {len(df)}")
overlap[['Start', 'Duration_min']]
```
